## Inheritances



```python
class person:   #class name

    def __init__(self,name,age):
        self.name = name
        self.age = age

    def introduce(self):
        print(f"Hi, I am {self.name} and I am {self.age} years old")
        
```

```python
p1 =person ('Thamn',34)
```

```python
class employee(person):

    def __init__(self, company):
        self.company = company

    def work(self):
        print(f" Hi, I am working at a{self.company}")
        
```

```python
emp2 = employee('wrop')
```

```python
emp2.company
```

```python
class employee(person):

    def __init__(self, name, age,company):
        super().__init__(name, age)
        self.company = company

    def work(self):
     super().introduce()
     print(f"Hi, I am working at {self.company}")
```

```python
emp2 = employee('Suman',45,'wrop')
```

```python
emp2.age
```

```python
emp2.company
```

```python
emp2.work()
```

```python
emp2.introduce()
```

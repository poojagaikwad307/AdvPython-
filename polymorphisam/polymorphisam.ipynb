#### User defined polymorphism scenario

```python
class India:

    def __init__(self):
        pass
    def capital(self):
        print("Delhi is the capital of India")

    def language(self):
        print("Hight and English are widely spoken language of India")
```

```python
class USA:

    def __init__(self):
        pass

    def capital(self):
        print("Washington D.C is the capital of USA")

    def language(self):
        print("English is the sidely spoken language of USA")
        
```

```python
ctry1 = India()
ctry2 = USA()
```

```python
class handale_missing_cont:

    def __init__(self,data):
        self.data = data

    def replace(self):
        print(f"Missing values before replace operation: {(self.data).isna().sum()}")
        md = self.data.mode()[0]
        (self.data).fillna(md,inplace=True)
        print(f"Missing values after replace operation: {(self.data).isna().sum()}")
        print(f"Missing values are replaced with {md}")
```

```python
cont1 = handale_missing_cont(cars['Luggage.room'])
```
